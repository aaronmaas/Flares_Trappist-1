## Effect of limb darkening:

We also should consider that the temperature of the star is smaller at the limb. So if a significant fraciton of the flare is close to the limb, the flare temperature would actually be overestimated! Since limb darkening leads to colder regions. 



The surface gravity of Trappist-1:

\begin{align}
F = m M G /R^2 = m \cdot g
\end{align}

Thus, 

\begin{align}
g_\mathrm{TRAP} = \frac{MG}{R^2} 
\end{align}

(For finding an appropriate model for TRAPPIST-1, for LD coeffcientes) 

Limb darkening can be expressed by  (Clare et al. 2002), 

\begin{align}
\frac{I(µ)}{I(1)} = 1 − a_1 \cdot (1 − µ^{1/2}) − a_2 \cdot (1 − µ) −a3 \cdot (1 − µ^{3/2}) - a4 \cdot (1 − µ^{2})
\end{align}

where $\mu = cos(\theta)$ and $\theta$ denotes the angle off the line of sight measured from the center of the stellar disk. $a_i$ are the different limb-darkening coefficients, $I(\mu)$ is the intensity of the star at $\mu$ and $I(1)$ ist the intensity at the center. 

```python
#limb dark coeffs for g-band

#Calculation of the limb darkening in the different Muscat passbands
import numpy as np 
ld_coeffs = np.genfromtxt("../data/TRAPPIST-1_limb_darkening/ldarkening_TRAPPIST-model.txt") #limb darkening coeffc a1, a2, a3, a4 for u,g,r,i,z filters (sloan)
#only griz not u band
a1 = ld_coeffs[0:1,6:10]
a2 = ld_coeffs[1:2,6:10]
a3 = ld_coeffs[2:3,6:10]
a4 = ld_coeffs[3:4,6:10]
mu_a = np.linspace(0,90,100) #angles
mu = np.cos(mu_a*np.pi/180) 

def limb_darkening(a1,a2,a3,a4,mu):
    intensity = []
    for i in range(len(a1)): 
        for j in range(len(mu)):
            intensity.append(1 - a1[i]*(1- (mu[j])**0.5)- a2[i]*(1-mu[j]) - a3[i]*(1-mu[j]**1.5 ) - a4[i]*(1-mu[j]**2 ))
            
    return intensity

ld60 = np.asarray(limb_darkening(a1,a2,a3,a4,mu))
```

```python
def find_nearest(array1, array2, value):
        idx = (np.abs(array1-value)).argmin()
        return array2[idx], idx
```

Thinking on the impact to our calculation the effective temperature of the TRAPPIST-1 decreases when approaching the edge of the star, due to:

\begin{align}
\int 0.5 B_\lambda \propto 0.5 T_\mathrm{eff}^4 = (T_\mathrm{eff} \cdot 0.841)^4 
\end{align}
