## Model for fitting 

The amount of radiant energy per unit time emitted by a blackbody of Temperature $T$ and surface are $dA$ into a solid angled $d\Omega = sin(\theta)d\theta d\phi$ is given by,

\begin{align}
d \tilde{l}_\lambda = B_\lambda(T) dA cos(\theta) d\Omega 
\end{align}

(Introduction to astrophysics Carrol and Ostlie S.95)

where each small patch of the blackbody $dA$ emits blackbody radiation into the solid angle. Now we have to integrate over only one side of the star. Thus, we integrate over $d\phi = [0,2\pi]$, $d\theta = [0,\pi/2]$ and only over the half of the surface area.

Performing the integral we obtain, 

\begin{align}
L_\lambda = \int d\phi \int cos(\theta) sin(\theta) d\theta \int B_\lambda dA = \int d\phi \int -\mu d\mu \int B_\lambda dA
\end{align}

where $\mu = cos(\theta)$. The integral over the half of the surface area gives us a factor of $4\pi R_*^2$ because, 

\begin{align}
\int B_\lambda dA = \int_0^{2 \pi} \int_{-\pi/2}^{\pi/2} B_\lambda R_*^2 sin(\theta) d\theta d\phi = 4\pi R_*^2 B_\lambda
\end{align}

The second integral $\int_0^{\pi/2} cos(\theta)sin(\theta) d\theta = \int_1^{0} -\mu d\mu = \int_{0}^1 \mu d\mu =  0.5$ and the third part $\int_0^{2\pi} d\phi = 2\pi$. Thus, we calculate the luminosity that is emitted into a half space of the visible half of the star. 

\begin{align}
L_\lambda = \int_0^{2\pi} d\phi \int_{0}^{\pi/2} cos(\theta) sin(\theta) d\theta \int_{dA} B_\lambda(T_*) d\tilde{A} 
          = \frac{4\pi}{2}  2\pi R_*^2 B_\lambda(T_*) = 4\pi^2 R_*^2 B_\lambda(T_*)
\end{align}

The brightness, i.e. flux density at the observer, $l_\lambda$ takes the Luminosity distance into account with the distance modulus,

\begin{align}
l_\lambda = \frac{L_\lambda}{4\pi d^2} = \frac{R_*^2}{d^2} B_\lambda (T_*) \pi
\end{align}


Then the brightness radiated from the stellar surface at an observers distance $d$ is,

\begin{align}
l_\lambda = \frac{R_*^2}{d^2} \cdot \frac{1}{\exp{\frac{hc}{k_b T_* \lambda}} - 1}  \cdot \frac{2hc^2}{\lambda^5} \cdot \pi
\end{align}

Now we can think of the flare brightness. The flare is associted to the total covered surface area from the flare, which is a projected semi sphere. We make one asumption here, the projected flare area is totally on the visbile side of the star. Before we proceed we have to make a distinction by cases. Either the flare area is optically thick or optically thin. 

First let us think of the optical thick case. The underlying flux from the stellar surface is suppressed here, such that,

\begin{align}
l_{flare} = \frac{R_{flare}^2}{d^2} \cdot B_\lambda(T_{flare}) \cdot \pi = \frac{(a \cdot R_{*})^2}{d^2} \cdot B_\lambda(T_{flare}) \cdot \pi = \tilde{a} \cdot l_\lambda (T_{flare})
\end{align} 

where $\tilde{a} = \sqrt{a}$ indicates the square root of the projected radius fraction of the star in the flaring region. 

The above euqation leads to the total observed brightness in the optically thick case, 

\begin{align}
l_{total} = l_{flare} + l_{star,reduced} = \tilde{a} \cdot l_\lambda(T_{flare}) + (1 - \tilde{a})\cdot l_\lambda(T_*)
\end{align}

Argument for optically thick (rephrase). The plasma at the photosphere is hotter than the plasma of the star! Thus, we cannot even see the same blackbody contribution from the star in the surface region associated with the flare.

Since we get out of the davenport template $l_{flare}$ seperated from the stellar brighness, we can fit independently for it. 

\begin{align}
l_{flare} = \tilde{a} \cdot (l_\lambda(T_{flare}) - l_\lambda(T_*))
\end{align}


The optical thin approach would yield, 

\begin{align}
    l_{flare} = \tilde{a} \cdot (l_\lambda(T_{flare}) + l_\lambda(T_*))
\end{align}
